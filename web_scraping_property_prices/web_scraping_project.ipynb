```python
import requests
from bs4 import BeautifulSoup

r = requests.get('https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/')
c = r.content

soup=BeautifulSoup(c,'html.parser')

all=soup.find_all('div',{'class':'propertyRow'})

for item in all:
    print(item.find_all('h4')[0].text.replace('\n','').replace(' ',''))
    
```

```python
for item in all:
    print(item.find_all('span',{'class':'propAddressCollapse'})[0].text)
    print(item.find_all('span',{'class':'propAddressCollapse'})[1].text)
    print('\n')
    
```

```python
l = []
for item in all:
    d={}
    d['Price'] = item.find('h4').text.replace('\n','').replace(' ','')
    d['Address'] = item.find_all('span',{'class':'propAddressCollapse'})[0].text
    try:
        d['Locality'] = item.find_all('span',{'class':'propAddressCollapse'})[1].text
    except:
        d['Locality'] = 'None'
    try:
        d['Beds'] = item.find('span',{'class':'infoBed'}).find('b').text
    except:
         d['Beds'] ='None'
    try:
        d['Baths'] = item.find('span',{'class':'infoValueFullBath'}).find('b').text
    except:
        d['Baths'] = 'None'
    try:
        d['Area'] = item.find('span',{'class':'infoSqFt'}).find('b').text
    except:
        d['Area'] ='None'
    for column_group in item.find_all('div',{'class':'columnGroup'}):
        for feature_group, feature_name in zip(column_group.find_all('span',{'class':'featureGroup'}),column_group.find_all('span',{'class':'featureName'})):
                if 'Lot Size' in feature_group.text:
                      d['Lot Size'] = feature_name.text    
    l.append(d)
 
```

```python
import pandas
df=pandas.DataFrame(l)
df
```

```python
page_num = soup.find_all('a',{'class':'Page'})
page_num=int(page_num[-1].text)*10
```

```python
base_url = 'https://pythonizing.github.io/data/real-estate/rock-springs-wy/LCWYROCKSPRINGS/t=0&s='
l = []
for page in range(0,page_num,10):
    r = requests.get(base_url+str(page))
    c = r.content
    soup = BeautifulSoup(c,'html.parser')
    all=soup.find_all('div',{'class':'propertyRow'})
    for item in all:
        d={}
        d['Price'] = item.find('h4').text.replace('\n','').replace(' ','')
        d['Address'] = item.find_all('span',{'class':'propAddressCollapse'})[0].text
        try:
            d['Locality'] = item.find_all('span',{'class':'propAddressCollapse'})[1].text
        except:
            d['Locality'] = 'None'
        try:
            d['Beds'] = item.find('span',{'class':'infoBed'}).find('b').text
        except:
             d['Beds'] ='None'
        try:
            d['Baths'] = item.find('span',{'class':'infoValueFullBath'}).find('b').text
        except:
            d['Baths'] = 'None'
        try:
            d['Area'] = item.find('span',{'class':'infoSqFt'}).find('b').text
        except:
            d['Area'] ='None'
        for column_group in item.find_all('div',{'class':'columnGroup'}):
            for feature_group, feature_name in zip(column_group.find_all('span',{'class':'featureGroup'}),column_group.find_all('span',{'class':'featureName'})):
                    if 'Lot Size' in feature_group.text:
                          d['Lot Size'] = feature_name.text    
        l.append(d)
    
```

```python
print(l)
```

```python
import pandas
df=pandas.DataFrame(l)
df['Lot Size'] = df['Lot Size'].fillna('None')
df = df[['Address','Locality','Beds','Baths','Area','Lot Size','Price']]
df.to_csv('Output.csv')
df
```
